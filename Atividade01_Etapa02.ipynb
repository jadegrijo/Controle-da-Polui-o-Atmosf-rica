Numpy Arrays

```python
import numpy as np

# Criando um vetor de dados aleatórios
data = np.random.rand(1000,2) #cria matriz de 1000 linhas e 2 colunas

# Verificando o tamanho do vetor
print(data.shape) #imprime o formato da matriz (1000,2)
print(np.size(data)) #imprime o tamanho da matriz, ou seja, quantos elementos tem dentro da matriz, no caso, 2x1000=2000

# Print de uma amostra dos dados
print(data[0:10]) #imprime as primeiras 10 linhas da matriz 
print(data[0:10].shape) #mostra o formato das primeiras 10 linhas da matriz
```

```python
# Usando Monte-Carlo
dataMax = 1000 #valor maximo de dados
dataMin = 10 #valor mínimo de dado
data = (dataMax - dataMin)*data + dataMin #garante que os dados comecem na datamin 
print(data[0:10]) #print da matriz
```

```python
import matplotlib.pyplot as plt #importa a biblioteca Matplotlib para criar os gráficos

fig, ax = plt.subplots(3) #cria 3 gáficos e armazena na variavel fig e os subgráficos na matriz ax
ax[0].plot(data) #plota dados brutos
ax[1].plot(data[:,0]) #plota os dados da primeira coluna na segunda subplotagem
ax[2].plot(data[:,1]) #plota dados da segunda coluna na terceira subplotagem
print(type(data)) 
```

```python
# Criando um dado 2D 
data3d = np.sin(np.random.rand(1000,1000)) #cria matriz 3D de senos aleatórios. Matriz 1000x1000
print(data3d) 3print da matriz
```

```python
# Plotando um dado 3D
fig2, ax2 = plt.subplots() #cria uma figura
ax2.pcolor(data3d) #plota os dados 3D com cores
```

```python
# Criando vertor temporal
dtime = np.arange(np.datetime64("2022-01-01"), np.datetime64("2023-01-01"), np.timedelta64(1, "h")) #Gera um array de datas começando em 2022-01-0 e acabando em 2023-01-01. Especifica o intervalo de 1h 
print(dtime[1:10]) #printa os dados começando do segundo elemento ate o nono
print(dtime.shape)
print(type(dtime))
```
